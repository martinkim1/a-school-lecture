```python
import pandas as pd
import matplotlib.pyplot as plt

# matplotlib에서 한글을 사용하기 위한 설정
plt.rc('font', family='Malgun Gothic')

# xlsx 파일을 DataFrame으로 읽어오기
df = pd.read_excel('취업 및 창업 의사.xlsx', header=[0, 1], index_col=[0, 1])

# '청년15~39세'의 데이터만 선택
data = df.loc['청년15~39세', '소계']

# 의사별 비율을 리스트로 변환
reasons = df.columns.tolist()
values = data.values.tolist()

# 파이 차트 그리기
plt.pie(values, labels=reasons, autopct='%1.1f%%')
plt.title('청년(15~39세) 취업 및 창업 의사')
plt.show()
```
